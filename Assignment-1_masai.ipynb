```python
import pandas as pd

# Load the data from the 'bookseller.csv' file
file_path = '/content/jTDOZrIDkLZ3MDUS.csv'  # Assuming this is the file location
data = pd.read_csv(file_path)

# Extract the column that contains the number of books sold per day
books_sold = data['Number of Books Sold']

# Question 1: Probability that more than (or equal to) 96 books are sold
prob_more_than_96 = (books_sold >= 96).sum() / len(books_sold)

# Question 2: Probability that less than (or equal to) 92 books are sold
prob_less_than_92 = (books_sold <= 92).sum() / len(books_sold)

# Output the probabilities
print(f"Probability that more than or equal to 96 books are sold: {prob_more_than_96}")
print(f"Probability that less than or equal to 92 books are sold: {prob_less_than_92}")
```

```python
import pandas as pd
debugging_data=pd.read_csv('/content/20sMGjG0yMJtVQGi.csv')
# 1. Probability that debugging requires less than 3 hours
total_bugs = len(debugging_data)
bugs_less_than_3_hours = len(debugging_data[debugging_data['Time Taken to fix the bug'] < 3])
probability_less_than_3_hours = bugs_less_than_3_hours / total_bugs

# 2. Probability that debugging requires more than 2 hours
bugs_more_than_2_hours = len(debugging_data[debugging_data['Time Taken to fix the bug'] > 2])
probability_more_than_2_hours = bugs_more_than_2_hours / total_bugs

# 3. 50th percentile (median) of debugging time
percentile_50 = debugging_data['Time Taken to fix the bug'].median()

# Output the results
print(f"Probability that debugging takes less than 3 hours: {probability_less_than_3_hours}")
print(f"Probability that debugging takes more than 2 hours: {probability_more_than_2_hours}")
print(f"50th percentile (median) of debugging time: {percentile_50}")
```
